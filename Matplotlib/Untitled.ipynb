#### Matplotlib Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("Hello Coder's")
plt.ylabel("Y label")
plt.xlabel("X label")
```

#### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x **2, x **2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

#### Object Oriented Way

```python
# Selecting canvas
fig = plt.figure(figsize=(10,8), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3, 0.1, 0.4, 0.4])
axis2.plot(x,y)
```

#### Types of Plots

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data) # only x value will be given
```

```python
plt.boxplot(x)
```

#### Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.savefig("basicPlot.png")
```
